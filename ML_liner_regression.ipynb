```python
import random
import matplotlib.pyplot as plt 
import pandas as pd 
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('iphone_price.csv')
print(df)
```

```python
x=df[['version']]
y=df[['price']]

xtrain,xtast,ytrain,ytest=train_test_split(x,y,test_size=.30,random_state=1)
reg=LinearRegression()
reg.fit(xtrain,ytrain)
```

```python
color=[random.randint(1,10) for _ in range(len(df['version']))]
size=[random.randint(100,500) for _ in range(len(df['version']))]

plt.scatter(df['version'],df['price'],c=color,s=size,alpha=0.5,cmap='jet')
plt.xlabel('iphone version',color='darkred')
plt.ylabel("iphone price",color="darkred")
plt.title("iphone price chat",color='hotpink')
plt.plot(df.version,reg.predict(df[['version']]))
plt.colorbar()
plt.tight_layout()
plt.show()
```

```python
import random
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('iphone_price.csv')
print(df) 
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(df[['version']],df[['price']],random_state=1)
```

```python
reg=LinearRegression()
reg.fit(xtrain,ytrain)
```

```python
reg.predict([[15]])
```

```python
color=[random.randint(1,10) for _ in range(len(df['version']))]
size=[random.randint(100,500) for _ in range(len(df['version']))]

plt.scatter(df.version,df.price,c=color,s=size,cmap='jet',alpha=0.5)
plt.xlabel('iphone version',color='darkred')
plt.ylabel('iphone price',color='darkred')
plt.title("iphone price and version chat",color='hotpink')
plt.plot(df.version,reg.predict(df[['version']]))
plt.colorbar()
plt.grid(color='blue',ls='--',lw=0.5)
plt.show()
```
